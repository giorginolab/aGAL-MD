## RMSD whole protein - CA

```python
# Tutte le cartelle
structure = ['apo', 'apo_N215S', 'apo_R301Q', 'DGJ', 'DGJ_N215S', 'DGJ_R301Q']
replica = ['_1', '_2', '_3']

for s in structure:
    for r in replica:
        topology = f'../3GXT_reglyco/{s}{r}/equilibration/structure.pdb'
        trajectory = f'../3GXT_reglyco/{s}{r}/production/output.xtc'
        
        rmsd(f'{s}{r}_CA', topology, trajectory, refmol = topology, trajrmsdstr='protein and name CA',trajalnstr='protein',refalnstr='protein', refrmsdstr = 'protein and name CA', title = True)
        
```

```python

fig, ax = plt.subplots(figsize=(10, 5), ncols=2, sharey=True)

structure = ['apo', 'apo_N215S', 'apo_R301Q', 
             'DGJ', 'DGJ_N215S', 'DGJ_R301Q']
replicas = ['_1', '_2', '_3']

for s in structure:
    for r in replicas:
        data = pd.read_csv(f'{s}{r}_CA.csv')
        label = f"{s}{r}"   # include replica number in label
        if s.startswith('apo'):
            ax[0].plot(data['time'], data['rmsd'], 
                       linestyle="-", linewidth=0.8, label=label, alpha=0.7)
        else:
            ax[1].plot(data['time'], data['rmsd'], 
                       linestyle="-", linewidth=0.8, label=label, alpha=0.7)

for a, title in zip(ax, ["Apo systems", "DGJ systems"]):
    a.set_xlabel("Time (ns)")
    a.set_ylabel("RMSD ()")
    a.set_title(title)
    a.legend(fontsize=7)

plt.tight_layout()
plt.show()
```

## RMSF whole protein - CA

```python
# Tutte le cartelle
structure = ['apo', 'apo_N215S', 'apo_R301Q', 'DGJ', 'DGJ_N215S', 'DGJ_R301Q']
replica = ['_1', '_2', '_3']

for s in structure:
    for r in replica:
        topology = f'../3GXT_reglyco/{s}{r}/equilibration/structure.pdb'
        trajectory = f'../3GXT_reglyco/{s}{r}/production/output.xtc'
        
        rmsf(f'{s}{r}_CA', topology, trajectory, atomsel = 'protein and name CA',refmol = topology, trajalnsel='protein and name CA', refalnsel='protein and name CA', title = True)
        
```

```python
structures = ['apo', 'apo_N215S', 'apo_R301Q',
              'DGJ', 'DGJ_N215S', 'DGJ_R301Q']
replicas = ['_1', '_2', '_3']

fig, axes = plt.subplots(2, 3, figsize=(20, 11), sharey=True)
axes = axes.flatten()

cmap = plt.cm.Paired
colors = [cmap(i) for i in range(6)]
replica_colors = {
    '_1A': colors[0], '_1B': colors[1],
    '_2A': colors[2], '_2B': colors[3],
    '_3A': colors[4], '_3B': colors[5],
}

for i, s in enumerate(structures):
    ax = axes[i]
    for r in replicas:
        data = pd.read_csv(f'../results/{s}{r}_CA_rmsf.csv')
        chainA = data[data['chain'] == 'A']
        chainB = data[data['chain'] == 'B']
        ax.plot(chainA['resid'], chainA['rmsf'],
                linestyle="-", linewidth=0.8, label=f"{r} - A", alpha=0.8,color=replica_colors[f"{r}A"])
        ax.plot(chainB['resid'], chainB['rmsf'],
                linestyle="-", linewidth=0.8, label=f"{r} - B", alpha=0.8,color=replica_colors[f"{r}B"])
    # Glycosylation sites
    first = True
    for marker_resid in [139, 192, 215]:
        ax.axvline(marker_resid, color="black", linestyle="-.", alpha=0.7,
                   label='Glycosylation site' if first else "")
        ymax = ax.get_ylim()[1]
        ax.text(marker_resid + 0.5, 0.5, str(marker_resid),
                va='center', ha='left', fontsize=7, color='black')
        first = False

    step = 10
    unique_resid = np.unique(data['resid'])
    all_ticks = sorted(set(list(unique_resid[::step])))
    ax.set_xticks(all_ticks)
    ax.set_xticklabels(all_ticks, rotation=45, fontsize=7)

    ax.set_xlim(data['resid'].min() - 0.5, data['resid'].max() + 0.5)
    ax.set_title(s, fontsize=12)

    if i % 3 == 0:
        ax.set_ylabel("RMSF ()")
    ax.set_xlabel("Residue ID")

handles, labels = ax.get_legend_handles_labels()
fig.legend(handles, labels, loc="upper center", ncol=2, fontsize=8)

plt.tight_layout()
plt.savefig("../results/rmsf_all_structures.pdf", dpi=300)
plt.show()
```
